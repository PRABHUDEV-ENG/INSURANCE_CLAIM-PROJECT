```python
import pandas as pd
import seaborn as sb
df=pd.read_csv(r"C:\Users\K.PRABHU\Downloads\dsp lab\insurance claim.csv")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.head()
```

```python
#distribution of claim status
#histoplot
plt.figure(figsize=(8,6))
sns.histplot(df['claim_status'], bins=30, kde=True)
plt.title("Distribution of Claim status")
plt.show()
```

```python
# 1. Line Plot – Subscription Length over Customer Age
sns.lineplot(x='customer_age', y='subscription_length', data=df)
plt.title("Line Plot: Subscription Length vs Customer Age")
plt.show()
```

```python
# 2. Area Plot – Vehicle Age Distribution
df['vehicle_age'].value_counts().sort_index().plot(kind='area', alpha=0.5)
plt.title("Area Plot: Vehicle Age Distribution")
plt.show()
```

```python
# 3. Histogram – Customer Age
sns.histplot(df['customer_age'], bins=30, kde=True)
plt.title("Histogram: Customer Age Distribution")
plt.show()
```

```python
# 4. Bar Chart – Average Subscription Length by Segment
sns.barplot(x='segment', y='subscription_length', data=df)
plt.title("Bar Chart: Subscription Length by Segment")
plt.show()
```

```python
# 5. Pie Chart – Fuel Type Distribution
df['fuel_type'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Pie Chart: Fuel Type Distribution")
plt.ylabel("")
plt.show()
```

```python
# 6. Box Plot – Subscription Length by Region Density
sns.boxplot(x='region_density', y='subscription_length', data=df)
plt.title("Box Plot: Subscription Length by Region Density")
plt.show()
```

```python
# 7. Scatter Plot – Vehicle Age vs Subscription Length
sns.scatterplot(x='vehicle_age', y='subscription_length', data=df)
plt.title("Scatter Plot: Vehicle Age vs Subscription Length")
plt.show()
```

```python
# 8. Bubble Plot – Customer Age vs Subscription Length (bubble size = vehicle_age)
plt.scatter(df['customer_age'], df['subscription_length'], 
            s=df['vehicle_age']*10, alpha=0.5)
plt.title("Bubble Plot: Customer Age vs Subscription Length")
plt.xlabel("Customer Age")
plt.ylabel("Subscription Length")
plt.show()
```

```python
# 9. Waffle Chart – Transmission Type Distribution
data = df['transmission_type'].value_counts().to_dict()
fig = plt.figure(
    FigureClass=pywaffle.Waffle, 
    rows=5, values=data, legend={'loc':'upper left', 'bbox_to_anchor':(1,1)}
)
plt.title("Waffle Chart: Transmission Type Distribution")
plt.show()
```

```python
# 10. Regression Plot – Customer Age vs Subscription Length
sns.regplot(x='customer_age', y='subscription_length', data=df)
plt.title("Regression Plot: Customer Age vs Subscription Length")
plt.show()
```

```python
# 11. Heatmap – Correlation of Numeric Features
numeric_df = df.select_dtypes(include=['int64','float64'])
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title("Heatmap: Numeric Feature Correlations")
plt.show()
```

```python
# 12. Violin Plot – Subscription Length by Segment
sns.violinplot(x='segment', y='subscription_length', data=df)
plt.title("Violin Plot: Subscription Length by Segment")
plt.show()
```

```python
# 13. Pair Plot – Customer Age, Vehicle Age, Subscription Length
sns.pairplot(df[['customer_age','vehicle_age','subscription_length']])
plt.suptitle("Pair Plot: Key Numeric Relationships", y=1.02)
plt.show()
```

```python
# 14. Stacked Bar Chart – Region Density vs Fuel Type
pd.crosstab(df['region_density'], df['fuel_type']).plot(kind='bar', stacked=True)
plt.title("Stacked Bar Chart: Region Density vs Fuel Type")
plt.show()
```
